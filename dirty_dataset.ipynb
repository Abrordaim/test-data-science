```python
import pandas as pd
import numpy as np

# Load dataset asli
df = pd.read_csv("user_behavior_dataset.csv")

# Copy untuk membuat versi kotor
raw = df.copy()

# ============================
# 1. Tambahkan Missing Values
# ============================
for col in raw.columns[:5]:
    idx = raw.sample(30).index
    raw.loc[idx, col] = np.nan

# ============================
# 2. Tambahkan Baris Duplikasi
# ============================
duplicates = raw.sample(20, random_state=42)
raw = pd.concat([raw, duplicates], ignore_index=True)

# ============================
# 3. Tambahkan Outlier Besar
# ============================
num_cols = raw.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    idx_outlier = raw.sample(5).index
    raw.loc[idx_outlier, col] = raw[col] * np.random.randint(20, 80)

# ============================
# 4. Buat Data Teks Jadi Berantakan
# ============================
if "Device_Model" in raw.columns:
    raw.loc[:15, "Device_Model"] = raw.loc[:15, "Device_Model"].astype(str).str.upper()

if "OS" in raw.columns:
    raw.loc[30:60, "OS"] = raw.loc[30:60, "OS"].astype(str).str.lower()

if "Gender" in raw.columns:
    raw.loc[70:100, "Gender"] = " Male"      # leading space
    raw.loc[120:150, "Gender"] = "female "   # trailing space

# ============================
# 5. Tambahkan Typo Buatan
# ============================
if "Device_Model" in raw.columns:
    raw.loc[5:10, "Device_Model"] = "SAMSNG-GLLXY"

# ============================
# 6. Simpan dataset hasil kotor 
# ============================
raw.to_csv("user_behavior_raw_dirty.csv", index=False)

print("Data mentah berhasil dibuat!")
```
